# income csv to dataframe

```python
import pandas as pd

income = pd.read_csv (r'kaggle_income.csv')
print (income)
```

```python
income_db = pd.DataFrame(income)
income_db
```

```python
for col in income_db.columns:
    print(col)
```

```python
clean_income_db = income_db.loc[:, ["State_Name", "City", "Zip_Code", "Mean", "Median", "Stdev", "Lat", "Lon"]]
clean_income_db
```

```python
clean_income_db = clean_income_db[(clean_income_db[['Mean', 'Median', 'Stdev']] != 0).all(axis=1)]
clean_income_db
```

# 2nd attempt at flattening json

```python
import pandas as pd
from pathlib import Path
import json
```

```python
p = Path(r'us_representation.json')

with p.open('r', encoding = 'utf-8') as f:
    data = json.loads(f.read())
    
officials = pd.json_normalize(data, record_path = '', meta = ['name', 'phone', 'party'])
```
